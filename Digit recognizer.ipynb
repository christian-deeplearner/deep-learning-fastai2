```python
def split_df(df):
    '''return a tuple (X, y) 
    
        X : the training inputs which is in (samples, height, width, channel) shape
        y : the label which is flatten
    '''
    y = df['label'].values.flatten()
    X = df.drop('label', axis=1).values
    X = X.reshape(X.shape[0], 28, 28)
    return (X,y)
```
